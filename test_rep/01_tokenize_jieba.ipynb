### 1. 分词模式

```python
import jieba
```

```python
text = "小明毕业于北京大学计算机系"
```

```python
# 精确模式
print(jieba.lcut(text))
```

```python
# 全模式
print(jieba.lcut(text, cut_all=True))
```

```python
# 搜索引擎模式
print(jieba.lcut_for_search(text))
```

### 2. 自定义词典
